```python
import pandas as pd
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
JANUARY_YELLOW_2023 = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet'
FEBRUARY_YELLOW_2023 = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet'
```

```python
df = pd.read_parquet(JANUARY_YELLOW_2023)
```

```python
num_columns = len(df.columns)
num_columns
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds()/60)
df.duration.describe()
```

```python
df.duration.iloc[0].total_seconds
```

```python
((df.duration >= 1) & (df.duration <= 60)).mean()
```

```python
df = df[((df.duration >= 1) & (df.duration <= 60))]
```

```python
categorical = ['PULocationID','DOLocationID']
numerical = ['trip_distance']
```

```python
df[categorical] = df[categorical].astype(str)
```

```python
train_dicts = df[categorical].to_dict(orient='records')
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
X_train
```

```python
target = 'duration'
```

```python
y_train = df[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
root_mean_squared_error(y_train, y_pred)
```


VALIDATION
